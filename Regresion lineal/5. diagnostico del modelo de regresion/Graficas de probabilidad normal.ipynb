```r
library(alr4)
```

## Observaciones de distribucion Normal 

```r
qqPlot(rnorm(50), envelope=FALSE)
```

## Observaciones de distribucion T de Student (colas pesadas) 

```r
qqPlot(rt(50,df=2), envelope=FALSE)
```

## Observaciones de distribucion Uniforme (colas ligeras)

```r
qqPlot(runif(50), envelope=FALSE)
```

## Observaciones de distribucion Ji-cuadrada (asimetria positiva)

```r
qqPlot(rchisq(50,df=2), envelope=FALSE)
```

## Prueba Shapiro-Wilk para probar la normalidad

```r
shapiro.test(rt(50,df=2))
```
